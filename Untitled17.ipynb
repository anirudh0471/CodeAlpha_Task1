```python
from googletrans import Translator, LANGUAGES
import ipywidgets as widgets
from IPython.display import display, clear_output
from gtts import gTTS
import pyperclip
import os
```

```python
translator = Translator()
language_codes = list(LANGUAGES.keys())
language_names = [LANGUAGES[code].title() for code in language_codes]

source_dropdown = widgets.Dropdown(
    options=list(zip(language_names, language_codes)),
    description='From:',
    value='en',
    layout=widgets.Layout(width='50%')
)

target_dropdown = widgets.Dropdown(
    options=list(zip(language_names, language_codes)),
    description='To:',
    value='es',
    layout=widgets.Layout(width='50%')
)

text_input = widgets.Textarea(
    value='Hello, how are you?',
    placeholder='Type something to translate...',
    description='Input:',
    layout=widgets.Layout(width='100%', height='100px')
)

translate_button = widgets.Button(description='Translate', button_style='success')
copy_button = widgets.Button(description='Copy', button_style='info')
tts_button = widgets.Button(description='Speak', button_style='primary')

output_text = widgets.Textarea(
    placeholder='Translation will appear here...',
    description='Output:',
    layout=widgets.Layout(width='100%', height='100px'),
    disabled=True
)
```

```python
def translate_text(b):
    clear_output(wait=True)
    display(ui)
    try:
        src = source_dropdown.value
        dest = target_dropdown.value
        text = text_input.value
        translated = translator.translate(text, src=src, dest=dest)
        output_text.value = translated.text
    except Exception as e:
        output_text.value = f"Error: {str(e)}"

def copy_to_clipboard(b):
    pyperclip.copy(output_text.value)

def speak_translation(b):
    tts = gTTS(text=output_text.value, lang=target_dropdown.value)
    tts.save("translation.mp3")
    os.system("mpg123 translation.mp3")
```

```python
translate_button.on_click(translate_text)
copy_button.on_click(copy_to_clipboard)
tts_button.on_click(speak_translation)
```

```python
manual_copy_label = widgets.Label("You can manually copy the translated text above.")
ui = widgets.VBox([
    source_dropdown,
    target_dropdown,
    text_input,
    translate_button,
    output_text,
    manual_copy_label,
    widgets.HBox([tts_button])
])
```
